```python
 import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns;import statistics as stat
import matplotlib as plt
import statistics
data=pd.read_csv('insurance.csv')
#reading data into working enviroment
data.head()

sns.set_style("darkgrid"); sns.set_palette(["#823A3A","#5ab4ac"]);sns.set_context('paper')
```

```python
print('Below is the number of smokers in the current dataset:') 
data['smoker'].value_counts()
```

```python
sns.pairplot(data, hue="smoker",corner=True)
```

```python
b = sns.lmplot(x="bmi",
               y="charges",
               col="sex",
               data=data,
               hue='smoker' )
b.fig.suptitle('Medical Costs & Body Mass Index by Smoker',y=1.09)
b.set(xlabel='Body Mass Index',ylabel= 'Medical Costs in Dollars')
```

```python


a=sns.relplot(x="age", 
                       y="charges",
                       hue='smoker',
                       data=data,
                       kind="line",
                       ci=True,
                       col='sex')
a.fig.suptitle("INSURANE COST OF SMOKERS and NON-SMOKERS VS AGE BY GENDER",y=1.09)
a.set(xlabel='Age',ylabel= 'Medical Costs in Dollars')

```

```python
sns.catplot(x='region',
            y='charges',
            hue='smoker',
            data=data,
            kind='swarm')
```

```python
b = sns.lmplot(x="bmi",
               y="charges",
               col="region",
               data=data,
               hue='smoker' )
b.fig.suptitle('Medical Costs & Body Mass Index by Smoker',y=1.09)
b.set(xlabel='Body Mass Index',ylabel= 'Medical Costs in Dollars')
```
